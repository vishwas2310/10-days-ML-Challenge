```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import country_converter as coco
import pycountry_convert as pc
```

# Process Data and Remove Days having more than 70% cases as 0

```python
data=pd.read_csv("time_series_2019-ncov-Confirmed.csv")
```

```python
data.head()
```

```python
columns=data.columns
for i in range(4,columns.size):
    total_count=data[columns[i]].size
    count=data[columns[i]].value_counts()
    zero_count=count.get(0)
    per=zero_count/total_count
    if(per>0.7):
        data=data.drop(columns[i],axis=1)
processed_data=data
```

```python
processed_data
```

# Plot data Country wise

```python
Country_wise=processed_data.drop('Lat',axis=1)
Country_wise=Country_wise.drop('Long',axis=1)
Country_wise=Country_wise.drop('Province/State',axis=1)
Country_wise=Country_wise.groupby('Country/Region')
```

```python
Processed=Country_wise.aggregate(np.sum)
Country_wise=Processed
```

```python
columns=Country_wise.columns
Country_wise=Country_wise.sort_values(columns[-1],ascending=False)
Country_wise=Country_wise.reset_index()
Country_wise.head()
```

```python
country=list(Country_wise['Country/Region'])
Total=list(Country_wise[columns[-1]])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(country[:10],Total[:10])
```

# Converting the Country Data to Continent

```python
Continent=[]
for i in country:
#     print(i)
    try:
        if(i=='Congo (Kinshasa)'):
            i="Congo"
        if(i=='Congo (Brazzaville)'):
            i="Congo"
        if(i=="Bahamas, The"):
            i="Bahamas"
        if(i=="Taiwan*"):
            i="Taiwan"
        country_code = pc.country_name_to_country_alpha2(i, cn_name_format="default")
    except:
        if(i=="Cote d'Ivoire"):
            i="CI"
        if(i=="Cruise Ship"):
            i="JP"
        country_code=i
    try:
        continent_name = pc.country_alpha2_to_continent_code(country_code)
    except:
        if(i=="Kosovo"):
            continent_name="EU"
        if(i=="Gambia, The"):
            continent_name="AF"
        if(i=="Holy See"):
            continent_name="EU"
        if(i=="East Timor"):
            continent_name="AS"
        if(i=="Timor-Leste"):
            continent_name="AS"
        if(i=="Korea, South"):
            continent_name="AS"
    Continent.append(continent_name)
```

```python
Total=list(Processed_without_sum.iloc[:,-1])
```

```python
df = pd.DataFrame( {'Continent':Continent, 'total':Total})
```

```python
Final=df.groupby('Continent')
```

```python
Final=Final.aggregate(np.sum)
```

```python
Final=Final.reset_index()
```

```python
Continent=Final['Continent']
Total=Final['total']
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(Continent,Total)
```
